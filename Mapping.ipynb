```python
import pandas as pd
import numpy as np
```

```python
Amnet = pd.read_excel('Amnet.xlsm',sheet_name='Display',skiprows=1)
Amnet.head(1)
```

```python
Amnet.rename(columns={'DCM Placement ID':'Placement ID','Net Media Cost (Local Currency)':'SpendCreative'}, inplace=True)
```

```python
Amnet.head(1)
```

```python
Amnet_pivot = pd.pivot_table(Amnet,index=['Placement ID'],values=['Publisher Impressions','SpendCreative'],aggfunc=[np.sum])
```

```python
Amnet_pivot = Amnet_pivot['sum']
```

```python
Amnet_pivot.head(1)
```

```python
Amnet_pivot['CPM'] =( Amnet_pivot['Publisher Impressions'] / Amnet_pivot['SpendCreative']) * 1000
Amnet_pivot.head()
```

```python
cost_dict = dict(zip(Amnet_pivot.index.tolist(),Amnet_pivot['CPM'].tolist()))
```

```python
DCM = pd.read_excel('DCM.xlsx')
DCM.head(1)
```

```python
DCM.columns
```

# Placement ID not present in Amnet

```python
set(DCM['Placement ID'].tolist()) - (set(Amnet_pivot.index.tolist()))
```

```python
DCM['Spend'] = DCM['Placement ID']
DCM['Spend'] = DCM['Spend'].map(cost_dict)
```

```python
DCM['Spend'] = DCM['Spend'] * DCM['Impressions']
```

```python
new_DCM = DCM[['Creative ID','Placement ID','Spend']]
```
